```python
# mergenando parcialmente as bases df_people e a df_campaing
df_parcial = pd.merge(df_people, df_campaing, on = 'ID', how = 'outer' )
```

```python
# mergeando todas a bases
df_total = pd.merge(df_parcial, df_purchase, on = 'ID', how = 'outer')
```

```python
# visualizando a base de dados após o merge de todos os arquivos
df_total.head()
```

```python
# criar uma coluna de idade AGE (considerando como referencia 2014)
ano_atual = 2014
df_total['Age'] = ano_atual - df_total['Year_Birth']
```

```python
# conferindo a criação da coluna AGE
df_total.head()
```

```python
# selecionando os respondentes (retirando os não respondentes)
colunas_camps = [
    'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5',
    'AcceptedCmp1', 'AcceptedCmp2', 'Complain'
]
df_total_respondentes = df_total.loc[(df_total[colunas_camps] == 0).all(axis=1)]
df_total_respondentes
```
